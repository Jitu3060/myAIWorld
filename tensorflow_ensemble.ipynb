```python
import sklearn
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
X,y = make_moons(n_samples=500,noise=0.30,random_state=42)
X_train,X_test, y_train, y_test = train_test_split(X,y,random_state=42)
```

```python
log_clf = LogisticRegression(solver="lbfgs",random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=100,random_state=42)
svm_clf = SVC(gamma="scale",random_state=42,probability=42)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
def build_nn():
    model = Sequential([Dense(50,activation='relu',input_shape = [2]),Dense(1,activation='sigmoid')])
    model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
import tensorflow as tf
from scikeras.wrappers import KerasClassifier
```

```python
keras_clf = KerasClassifier(model=build_nn,epochs=500,verbose=False)
```

```python
voting = VotingClassifier(estimators=[('lr',log_clf),('rf',rnd_clf),('svc',svm_clf),('keras',keras_clf)],voting='soft',flatten_transform=True)
```

```python
voting.fit(X_train,y_train)
```

```python
for clf in (log_clf,rnd_clf,svm_clf,keras_clf,voting):
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))
```
